```python
#imports
import random
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
## 1.
# Wykonać wykres funkcji g(x) = (x^2 - 4x + 3)^2
# z wykładu na przedziale [0,4]
# korzystając z matplotlib w notatniku Jupyter

def g(x):
    return pow((pow(x, 2) - 4*x + 3), 2)

x = np.linspace(0, 4, 50)
y = g(x)

plt.plot(x, y, color='red')
plt.show()
```

```python
## 2.
# Napisać funkcję dekodującą łańcuch 16 wartości binarnych
# (podawanych w tablicy Numpy), zawierający liczbę w
# naturalnym kodzie binarnym, do wartości dziesiętnej. 

# generujemy losowe wartości binarne
def rng(col=8, row=16): # dla col = ilosc znakow w ciagu; row = ilosc ciagow;
    arr = []
    for rows in range(row):
        n = ""
        for cols in range(col):
            var = random.randint(0, 1)
            n += str(var)
        arr.append(n)
    return arr

def list_to_dec(n): # dla n = ['0110...10', '1...0'] (list of strings)
    arr = []
    for var in n:
        r = 0
        for i in range(len(var)):
            r += pow(2, i) * int(var[::-1][i])
        arr.append(r)
        #print(f"{var} is: {r}") # for debugging
    return arr

list_to_dec(rng())
```

```python
## 3.
# Napisać funkcję dekodującą łańcuch 16 wartości binarnych
# (podawanych w tablicy Numpy) do wartości dziesiętnej z
# przedziału w taki sposób, że:
# * wartość 000...0(2) jest dekodowana do 0(10),
# * wartość 111...1(2) jest dekodowana do 1(10),
# * pozostałe wartości są rozmieszczone równomiernie w przedziale [0, 1].

def rng(col=8, row=16):
    arr = []
    for rows in range(row):
        n = ""
        for cols in range(col):
            var = random.randint(0, 1)
            n += str(var)
        arr.append(n)
    return arr

def get_max(n): # dla n = '101...1' (string)
    r = 0
    for i in range(len(n)):
        r += pow(2, i) * 1
    return r

def list_to_normalized(n): # dla n = ['0110...10', '1...0'] (list of strings)
    arr = []
    max_number = get_max(n[0]) # zakladamy ze wszystkie elementy są tej samej długości
    for var in n:
        r = 0
        for i in range(len(var)):
            r += pow(2, i) * int(var[::-1][i])
        arr.append(r / max_number)
    return arr

list_to_normalized(rng())
```

```python
## 4.
# Na podstawie powyższej funkcji utwórz zmodyfikowaną funkcję,
# w której łańcuch wartości binarnych jest dekodowany do
# wartości dziesiętnej z przedziału [0, 4]

def rng(col=8, row=16):
    arr = []
    for rows in range(row):
        n = ""
        for cols in range(col):
            var = random.randint(0, 1)
            n += str(var)
        arr.append(n)
    return arr

def get_max(n): # dla n = '101...1' (string)
    r = 0
    for i in range(len(n)):
        r += pow(2, i) * 1
    return r

def list_to_normalized(n, amplitude=4): # dla n = ['0110...10', '1...0'] (list of strings)
    arr = []
    max_number = get_max(n[0])
    for var in n:
        r = 0
        for i in range(len(var)):
            r += pow(2, i) * int(var[::-1][i])
        arr.append(amplitude * r / max_number)
    return arr

list_to_normalized(rng())
```

```python
## 6. Utwórz funkcję, która wylicza dla zadanej populacji
# przystosowanie (czyli wartość ) każdego chromosomu w populacji.

def f(x):
    return math.sin(x) + math.sin((10/3) * x)

def func_normalized(n, max_number, val=10, offset=0): # dla n = int
    return f(offset + ((n / max_number) * val))

def rng(col=8, row=16):
    arr = []
    for rows in range(row):
        n = ""
        for cols in range(col):
            var = random.randint(0, 1)
            n += str(var)
        arr.append(n)
    return arr

def get_max(n): # dla n = '101...1' (string)
    r = 0
    for i in range(len(n)):
        r += pow(2, i) * 1
    return r

def fit(n): # dla n = '011...0' (string)
    dec = 0
    for i in range(len(n)):
        dec += pow(2, i) * int(n[::-1][i])
    return func_normalized(dec, get_max(n))

for i in rng():
    print(f"Przystosowanie dla {i}: {fit(i)}")
```
