```python
## 1.
# Wykonać wykres funkcji g(x) = (x^2 - 4x + 3)^2
# z wykładu na przedziale [0,4]
# korzystając z matplotlib w notatniku Jupyter

def g(x):
    return pow((pow(x, 2) - 4*x + 3), 2)

x = np.linspace(0, 4, 50)
y = g(x)

plt.plot(x, y, color='red')
plt.show()
```

```python
## 2.
# Napisać funkcję dekodującą łańcuch 16 wartości binarnych
# (podawanych w tablicy Numpy), zawierający liczbę w
# naturalnym kodzie binarnym, do wartości dziesiętnej. 

# generujemy losowe wartości binarne
def rng(col=8, row=16): # dla col = ilosc znakow w ciagu; row = ilosc ciagow;
    arr = []
    for rows in range(row):
        n = ""
        for cols in range(col):
            var = random.randint(0, 1)
            n += str(var)
        arr.append(n)
    return arr

def list_to_dec(n): # dla n = ['0110...10', '1...0'] (list of strings)
    arr = []
    for var in n:
        r = 0
        for i in range(len(var)):
            r += pow(2, i) * int(var[::-1][i])
        arr.append(r)
        #print(f"{var} is: {r}") # for debugging
    return arr

list_to_dec(rng())
```

```python
## 3.
# Napisać funkcję dekodującą łańcuch 16 wartości binarnych
# (podawanych w tablicy Numpy) do wartości dziesiętnej z
# przedziału w taki sposób, że:
# * wartość 000...0(2) jest dekodowana do 0(10),
# * wartość 111...1(2) jest dekodowana do 1(10),
# * pozostałe wartości są rozmieszczone równomiernie w przedziale [0, 1].

def rng(col=8, row=16):
    arr = []
    for rows in range(row):
        n = ""
        for cols in range(col):
            var = random.randint(0, 1)
            n += str(var)
        arr.append(n)
    return arr

def get_max(n): # dla n = '101...1' (string)
    r = 0
    for i in range(len(n)):
        r += pow(2, i) * 1
    return r

def list_to_normalized(n): # dla n = ['0110...10', '1...0'] (list of strings)
    arr = []
    max_number = get_max(n[0]) # zakladamy ze wszystkie elementy są tej samej długości
    for var in n:
        r = 0
        for i in range(len(var)):
            r += pow(2, i) * int(var[::-1][i])
        arr.append(r / max_number)
    return arr

list_to_normalized(rng())
```

```python
## 4.
# Na podstawie powyższej funkcji utwórz zmodyfikowaną funkcję,
# w której łańcuch wartości binarnych jest dekodowany do
# wartości dziesiętnej z przedziału [0, 4]

def rng(col=8, row=16):
    arr = []
    for rows in range(row):
        n = ""
        for cols in range(col):
            var = random.randint(0, 1)
            n += str(var)
        arr.append(n)
    return arr

def get_max(n): # dla n = '101...1' (string)
    r = 0
    for i in range(len(n)):
        r += pow(2, i) * 1
    return r

def list_to_normalized(n, amplitude=4): # dla n = ['0110...10', '1...0'] (list of strings)
    arr = []
    max_number = get_max(n[0])
    for var in n:
        r = 0
        for i in range(len(var)):
            r += pow(2, i) * int(var[::-1][i])
        arr.append(amplitude * r / max_number)
    return arr

list_to_normalized(rng())
```

```python
## 9.
# Utwórz funkcję, która dla zadanej populacji oraz prawdopodobieństwa
# krzyżowania zwraca nową populację, dla której zastosowano krzyżowanie.

def rng(col=8, row=16):
    arr = []
    for rows in range(row):
        n = ""
        for cols in range(col):
            var = random.randint(0, 1)
            n += str(var)
        arr.append(n)
    return arr

def handle_crossing(n, p):
    if(random.randrange(0, 100) / 100 >= p):
        print("\nKrzyżowanie nie odbyło się.")
        return n #skipping
    else:
        print("\nKrzyżowanie...")
        pairs = []
        npairs = []
        remaining = n
        for p in range(len(remaining) // 2):
            pairs.append( [ remaining.pop(random.randrange(len(remaining))),
            remaining.pop(random.randrange(len(remaining))) ] )
        for p in pairs:
            index = random.randint(0, len(p[0]))
            products = [""+p[0][:index]+p[1][index:], ""+p[0][index:]+p[1][:index]]
            # naprawdę ładna wizualizacja, ale zapełnia całą konsolę, więc zostawiam jako komentarz
            #print(f"\nKrzyżowanie par:\n{p[0][:index]}|{p[0][index:]}\n{p[1][:index]}|{p[1][index:]}\nw pkt. {index}:\n{products[0]}\n{products[1]} - wyniki\n")
            npairs.append(products)
        res = []
        for i in npairs:
            res.append(i[0])
            res.append(i[1])
        return res

def run(times, crossing_prob = .5):
    list = rng()
    for n in range(times):
        print(f"\n//Gen {n}")
        for i in list:
            print(i)
        list = handle_crossing(list, crossing_prob)

run(2)
```

```python
## 11.
# Korzystając z funkcji z zadań 5-9, stwórz algorytm genetyczny
# rozwiązujący zadany problem optymalizacji. Uruchom algorytm,
# a następnie narysuj na wykresie funkcji punkty odpowiadające wszystkim 
# chromosomom z ostatniej populacji oraz (innym kolorem)
# najlepsze znalezione przez algorytm rozwiązanie.

def f(x):
    return math.sin(x) + math.sin((10/3) * x)

def func_normalized(n, max_number, val=10, offset=0): # dla n = int
    return f(offset + ((n / max_number) * val))

def dec(n): # dla n = '011...1' (string)
    r = 0
    for i in range(len(n)):
        r += pow(2, i) * int(n[::-1][i])
    return r

def dec_norm_list(n, val=10): # dla n = ['01..10', ...] (list)
    a = []
    nmax = get_max(n[0])
    for i in n:
        a.append(val * dec(i) / nmax)
    return a

def rng(col=8, row=16):
    arr = []
    for rows in range(row):
        n = ""
        for cols in range(col):
            var = random.randint(0, 1)
            n += str(var)
        arr.append(n)
    return arr

def get_max(n): # dla n = '101...1' (string)
    r = 0
    for i in range(len(n)):
        r += pow(2, i) * 1
    return r

def fit(n): # dla n = '011...0' (string)
    dec = 0
    for i in range(len(n)):
        dec += pow(2, i) * int(n[::-1][i])
    return func_normalized(dec, get_max(n))

def process_weights(p): # funkcja normalizuje wagi, żeby były kompatybilne ze sposobem przypisywania wag w numpy
    if(np.min(p) < 0):
        offset = abs(np.min(p))
        for i in range(len(p)):        
            p[i] = p[i] + offset
    p = np.array(p)
    p /= p.sum()
    return p
    

def choose_random(list, weights):
    a = []    
    weights = process_weights(weights)
    a = np.random.choice(list, len(list), p=weights)
    return a

def handle_crossing(n, p):
    if(random.randrange(0, 100) / 100 >= p):
        return n #skipping
    else:
        pairs = []
        npairs = []
        remaining = n.tolist()
        for p in range(len(remaining) // 2):
            pairs.append( [ remaining.pop(random.randrange(len(remaining))),
            remaining.pop(random.randrange(len(remaining))) ] )
        for p in pairs:
            index = random.randint(0, len(p[0]))
            products = [""+p[0][:index]+p[1][index:], ""+p[0][index:]+p[1][:index]]
            npairs.append(products)
        res = []
        for i in npairs:
            res.append(i[0])
            res.append(i[1])
        return res

def handle_mutation(n, p):
    nlist = []
    for i in n:
        new = ""
        for b in range(len(i)):
            if(random.randrange(0, 100) / 100 <= p):
                new += str(abs(int(i[b]) - 1))
            else:
                new += i[b]
        nlist.append(new)
    return nlist

def run(times, crossing_prob=.01, mutation_prob=.01):
    list = rng()
    x = np.linspace(0, 10, 200)
    y = np.sin(x) + np.sin((10 / 3) * x)
    plt.plot(x, y, color='red')

    best_score = [0, -float('inf')]
    
    for n in range(times):
        val = []
        x1 = dec_norm_list(list)
        y1 = []
        for i in range(len(list)):
            val.append(fit(list[i]))
            y1.append(fit(list[i]))
            if(fit(list[i]) > best_score[1]):
                best_score = [x1[i], fit(list[i])] 
        list = choose_random(list, val)
        list = handle_crossing(list, crossing_prob)
        list = handle_mutation(list, mutation_prob)

        color = 'gray' if n < (times - 1) else 'blue'
        if(n == 0):
            label = 'pozostałe generacje'
        elif(n == times - 1):
            label = 'najnowsza generacja'
        else:
            label = None
        sc = plt.scatter(x1, y1, color=color, label=label)
        
    plt.scatter(best_score[0], best_score[1], color='gold', label='najlepszy wynik')
    plt.legend()

# użycie:
# run(ilość_powtórzeń,
#     szansa_na_krzyżowanie [n * 100%],
#     szansa_na_mutację [n * 100%])
run(128, .01, .01)
```

```python
## 12. 
# Korzystając z reprezentacji binarnej napisz algorytm genetyczny
# dla problemu plecakowego z opisaną poniżej funkcją celu.

item_weights = [2, 5, 2, 3, 5]
item_values = [5, 4, 7, 1, 5]
penalty_coefficient = 10
cap = 13

def rng(col=5, row=5):
    arr = []
    for rows in range(row):
        n = ""
        for cols in range(col):
            var = random.randint(0, 1)
            n += str(var)
        arr.append(n)
    return arr

def f(x): # n = '11..01' (string)
    score = 0
    weight = 0
    for i in range(len(x)):
        score += item_values[i] * int(x[i])
        weight += item_weights[i] * int(x[i])
    if(score > cap):
        score -= (weight - cap) * penalty_coefficient
    return score

def process_weights(p):
    if(np.min(p) < 0):
        offset = abs(np.min(p))
        for i in range(len(p)):        
            p[i] = p[i] + offset
    p = np.array(p, dtype=float)
    p /= p.sum()
    return p
    

def choose_random(list, weights):
    a = []    
    weights = process_weights(weights)
    a = np.random.choice(list, len(list), p=weights)
    return a

def handle_crossing(n, p):
    if(random.randrange(0, 100) / 100 >= p):
        return n #skipping
    else:
        pairs = []
        npairs = []
        remaining = n.tolist()
        for p in range(len(remaining) // 2):
            pairs.append( [ remaining.pop(random.randrange(len(remaining))),
            remaining.pop(random.randrange(len(remaining))) ] )
        for p in pairs:
            index = random.randint(0, len(p[0]))
            products = [""+p[0][:index]+p[1][index:], ""+p[0][index:]+p[1][:index]]
            npairs.append(products)
        res = []
        for i in npairs:
            res.append(i[0])
            res.append(i[1])
        return res

def handle_mutation(n, p):
    nlist = []
    for i in n:
        new = ""
        for b in range(len(i)):
            if(random.randrange(0, 100) / 100 <= p):
                new += str(abs(int(i[b]) - 1))
            else:
                new += i[b]
        nlist.append(new)
    return nlist

def run(times, crossing_chance=.1, mutation_chance=.1):
    list = rng()

    for n in range(times):
        list_fit = []
        print(f"\nGeneracja {n+1}")
        for i in list:
            score = f(i)
            print(f"Przystosowanie dla {i}: {score}")
            list_fit.append(score)
        list = choose_random(list, list_fit)
        list = handle_crossing(list, crossing_chance)
        list = handle_mutation(list, mutation_chance)

run(5)
```

```python
## 13. 
# Zrealizuj algorytm genetyczny, w którym jako funkcja celu występuje wartość
# przedmiotów w plecaku, ale dla (niektórych) rozwiązań przekraczających
# ograniczenie na pojemność plecaka stosowana jest procedura naprawy

item_weights = [2, 5, 2, 3, 5]
item_values = [5, 4, 7, 1, 5]
penalty_coefficient = 10
cap = 13

def rng(col=5, row=5):
    arr = []
    for rows in range(row):
        n = ""
        for cols in range(col):
            var = random.randint(0, 1)
            n += str(var)
        arr.append(n)
    return arr

def f(x): # n = '11..01' (string)
    score = 0
    weight = 0
    for i in range(len(x)):
        score += item_values[i] * int(x[i])
        weight += item_weights[i] * int(x[i])
    if(score > cap):
        score -= (weight - cap) * penalty_coefficient
    return score

def process_weights(p):
    if(np.min(p) < 0):
        offset = abs(np.min(p))
        for i in range(len(p)):        
            p[i] = p[i] + offset
    p = np.array(p, dtype=float)
    p /= p.sum()
    return p
    

def choose_random(list, weights):
    a = []    
    weights = process_weights(weights)
    a = np.random.choice(list, len(list), p=weights)
    return a

def handle_crossing(n, p):
    if(random.randrange(0, 100) / 100 >= p):
        return n #skipping
    else:
        pairs = []
        npairs = []
        remaining = n.tolist()
        for p in range(len(remaining) // 2):
            pairs.append( [ remaining.pop(random.randrange(len(remaining))),
            remaining.pop(random.randrange(len(remaining))) ] )
        for p in pairs:
            index = random.randint(0, len(p[0]))
            products = [""+p[0][:index]+p[1][index:], ""+p[0][index:]+p[1][:index]]
            npairs.append(products)
        res = []
        for i in npairs:
            res.append(i[0])
            res.append(i[1])
        return res

def handle_mutation(n, p):
    nlist = []
    for i in n:
        new = ""
        for b in range(len(i)):
            if(random.randrange(0, 100) / 100 <= p):
                new += str(abs(int(i[b]) - 1))
            else:
                new += i[b]
        nlist.append(new)
    return nlist

def repair(n): # implementacja naprawy losowej (jesli dobrze zrozumialem zadanie)
    repaired = []
    for g in n:
        ind = list(g)
        weight = 0
        for i in range(len(ind)):
            weight += item_weights[i] * int(ind[i])
        while weight > cap:
            ones = []
            for i in range(len(ind)):
                if ind[i] == '1':
                    ones.append(i)
            if len(ones) == 0:
                break
            idx = random.choice(ones)
            ind[idx] = '0'
            weight -= item_weights[idx]
        repaired.append("".join(ind))
    return repaired

def run(times, crossing_chance=.1, mutation_chance=.1):
    list = rng()

    for n in range(times):
        list_fit = []
        print(f"\nGeneracja {n+1}")
        for i in list:
            score = f(i)
            print(f"Przystosowanie dla {i}: {score}")
            list_fit.append(score)
        list = choose_random(list, list_fit)
        list = handle_crossing(list, crossing_chance)
        list = handle_mutation(list, mutation_chance)
        list = repair(list)

run(5)
```
